```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import json
import ROOT

import random
#random.seed(1)
#np.random.seed(1)
#import os
#os.environ['PYTHONHASHSEED']=str(1)
#tf.random.set_seed(1)
```

```python
with open("data.txt") as f:
    tracks = json.load(f)

random.shuffle(tracks)
```

```python
train_perc = 0.6

x_train = np.array([tracks[ev][0] for ev in range(int(train_perc * len(tracks)))])
y_train = np.array([tracks[ev][1] for ev in range(int(train_perc * len(tracks)))])

x_valid = np.array([tracks[ev][0] for ev in range(int(train_perc * len(tracks)), len(tracks))])
y_valid = np.array([tracks[ev][1] for ev in range(int(train_perc * len(tracks)), len(tracks))])
```

```python
def create_simple_nn():  
  model = keras.Sequential()
  model.add(layers.InputLayer(64, dtype=tf.bool, name="Input_layer"))
  model.add(layers.Dense(10, activation='relu', name="Hidden_layer_2"))
  model.add(layers.Dense(5, activation='relu', name="Hidden_layer_3"))
  model.add(layers.Dense(2, activation='softmax', name="Output_layer"))
  model.add(layers.BatchNormalization(axis=1, renorm=True))

  return model
```

```python
def my_metric_fn(y_true, y_pred):
    def obtain(true, pred):
        ind_real = list(true).index(1)
        ind_true = list(pred).index(max(pred))
        if ind_real == ind_true:
            return float(1)
        else:
            return float(0)
    return np.sum([obtain(*elem) for elem in zip(y_true, y_pred)])
print(my_metric_fn([[0, 0, 1], [0, 0, 1]], [[0.5, 0.25, 1.], [0.5, 0.25, 1.]]))
```

```python
snn_model = create_simple_nn()
#snn_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=[my_metric_fn])
snn_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc', 'mse'])
```

```python
snn = snn_model.fit(x=x_train, y=y_train, batch_size=16, epochs=50, verbose=1, validation_data=(x_valid, y_valid), shuffle=True)
```

```python
classes = snn_model.predict(x_valid, batch_size=16)
```

```python
for real, valid in zip(y_valid, classes):
    print(real, valid)
```

```python
keys = ["straight", "random"]
d = {}

for x, real, valid in zip(x_valid, y_valid, classes):
    key = keys[list(real).index(1)]
    found = keys[list(valid).index(max(valid))]
    # print real, key, valid, found
    if key not in d:
        d[key] = {}
    if found not in d[key]:
        d[key][found] = []
    d[key][found].append(list(x))
```

```python
import tabulate
def confusion_matrix(d):
    table = []
    keys = ["straight", "random"]
    headers = ["True/Pred"] + keys
    for true_key in keys:
        line = [true_key]
        for pred_key in keys:
            line.append(len(d[true_key][pred_key]))
        table.append(line)
    print(tabulate.tabulate(table, headers=headers))
confusion_matrix(d)
```

```python
import hls4ml
from tensorflow.keras.models import load_model

model = load_model("my_model.h5")

config = hls4ml.utils.config_from_keras_model(model, granularity='name')

from pprint import pprint
#pprint(config)

config['LayerName']['Input_layer']['Precision']['result'] = 'ap_int<2>'
config['Model']['Strategy'] = 'Resource'


#hls_kk = hls4ml.templates.vivado_template.VivadoBackend(name='Vivado').create_initial_config(clock_period=10)
#pprint(hls_kk)

hls_model = hls4ml.converters.convert_from_keras_model(
    model, hls_config=config, output_dir='test_prj',
    part='xc7a35tcpg236-1', clock_period = 40)

hls_model.config.model_strategy = 'Resource'

hls_model.write()
hls_model.compile()
```
